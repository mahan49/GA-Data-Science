```python
import pandas as pd
import json

data = pd.read_csv("../../../lesson-13/code/dataset/stumbleupon.tsv", sep='\t')
data['title'] = data.boilerplate.map(lambda x: json.loads(x).get('title', ''))
data['body'] = data.boilerplate.map(lambda x: json.loads(x).get('body', ''))
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

bodies = data['body'].dropna().copy()
vectorizer = CountVectorizer(max_features=1000, binary=False, stop_words='english')
```

```python
X = vectorizer.fit_transform(bodies)#Does both fit and transform
```

```python
id2word = dict(enumerate(vectorizer.get_feature_names()))
```

```python
id2word # List of all of the data columns
```

```python
from gensim.models.ldamodel import LdaModel
from gensim.matutils import Sparse2Corpus
# First we convert our word-matrix into gensim's format
corpus = Sparse2Corpus(X, documents_columns = False)
# Then we fit an LDA model
lda_model = LdaModel(corpus=corpus, id2word=id2word, num_topics=15)
```

```python
num_topics = 5 #Change this number to get a number of meaningful topics
num_words_per_topic = 10
for ti, topic in enumerate(lda_model.show_topics(num_topics = num_topics, num_words = num_words_per_topic)):
    print("Topic: %d" % (ti))
    print (topic)
    print()
```

```python
from gensim.models.word2vec import Word2Vec
# Setup the body text
text = data.body.dropna().map(lambda x: x.split())
model = Word2Vec(text, size=100, window=5, min_count=5, workers=4)
```

```python
model.most_similar(positive=['cookie', 'brownie'])
```
